```python
import numpy as np
import matplotlib.pyplot as plt
def inverseCDF_fretchet(alpha, u):
    return (-np.log(u))**(-1/alpha)
```

```python
alpha = 2
n = 1000
u = np.random.uniform(size=n)
sample = inverseCDF_fretchet(alpha, u)

```

```python
plt.figure(figsize=(8, 6))
plt.hist(sample, bins=50, density=True, alpha=0.5, color='blue')
plt.title(f'Kernel Density Estimation of Fr√©chet Distribution (alpha={alpha})')
plt.xlabel('x')
plt.ylabel('Density')
x = np.linspace(inverseCDF_fretchet(0.001, alpha), inverseCDF_fretchet(0.999, alpha), 100)
plt.plot(x, inverseCDF_fretchet(x, alpha), 'r-', lw=2, label='PDF')
plt.legend()
plt.show()
```
